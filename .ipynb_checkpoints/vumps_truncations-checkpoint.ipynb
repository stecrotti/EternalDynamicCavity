```julia
using TensorTrains, LinearAlgebra, Random
import TensorTrains.UniformTensorTrains: InfiniteUniformTensorTrain, one_normalization, _eigen
using TensorKit, MPSKit
using Unzip, Statistics
```

```julia
Q = 4
m = 30
ds = 1:(m÷10):m+5
nsamples = 30;
```

```julia
function truncate_vumps(A::Array, d::Integer)
    Q = size(A, 2)
    m = size(A, 1)
    @assert size(A, 3) == m
    t = TensorMap(A,(ℝ^m ⊗ ℝ^Q), ℝ^m) # the same but as a type digestible by MPSKit.jl
    ψ₀ = InfiniteMPS([t])
    I = DenseMPO([MPSKit.add_util_leg(id(storagetype(MPSKit.site_type(ψ₀)), physicalspace(ψ₀, i)))
        for i in 1:length(ψ₀)])
    ψ = InfiniteMPS(ℝ^Q, ℝ^d) # approximate by MPS of bond dimension d
    alg = VUMPS(; tol_galerkin=1e-15, maxiter=100) # variational approximation algorithm
    ψ, = approximate(ψ, (I, ψ₀), alg)   # do the truncation
    ovl = abs(dot(ψ, ψ₀))
    B = reshape(only(ψ.AL).data, d, Q, d)
    return B, ovl
end
```

```julia
d = 10
A = complex.(rand(m, Q, m))     # matrix to be truncated. size is mxm, variable with 2 states
B_, ovl = truncate_vumps(A, d)
```

```julia
q = TensorTrains.InfiniteUniformTensorTrain(permutedims(B_, (1,3,2)))
A_ = reshape(A, m, Q, m)
p = TensorTrains.InfiniteUniformTensorTrain(permutedims(A_, (1,3,2)));
```

```julia
only(marginals(q))
```

```julia
errs_avg, errs_std, ovls_avg, ovls_std = map(ds) do d
    e,o = map(1:nsamples) do _
        A = complex.(rand(m, Q, m))     # matrix to be truncated. size is mxm, variable with 2 states
        B_, ovl = truncate_vumps(A, d)
        q = TensorTrains.InfiniteUniformTensorTrain(permutedims(B_, (1,3,2)))

        A_ = reshape(A, m, Q, m)
        p = TensorTrains.InfiniteUniformTensorTrain(permutedims(A_, (1,3,2)))

        err = maximum(abs, only(marginals(p)) - only(marginals(q)))
        err, ovl
    end |> unzip
    mean(e), std(e)/sqrt(nsamples), mean(o), std(o)/sqrt(nsamples)
end |> unzip
```

Infinite MPS $\psi$ with bond dimension $m$=30, truncated to $\tilde\psi$ with bond dimension $d$ via the variational VUMPS method.
Quantities of interest are:
- The overlap $o=\langle\tilde\psi|\psi\rangle$, the function minimized in the variational approach (keeping both MPS's normalized in the 2-norm)
- The distributions in our classical sense $p(\overline x)=\frac{\psi(\overline x)}{\sum_{\overline x} \psi(\overline x)}$ and the truncated one $\tilde p(\overline x)=\frac{\tilde\psi(\overline x)}{\sum_{\overline x} \psi(\overline x)}$, from which we compute the corresponding marginal distributions for a single site $p_i(x_i)$, $\tilde p_i(x_i)$

In this experiment, $\psi(\overline x)$ has entries $\sim U([0,1])$ and each $x_i$ is binary. 

Left plot: $d$ versus $|1-o|$ (points with zero overlap do not show becauuse of the log-scale)

Right plot: $d$ versus $\max\limits_{x^i} |p_i(x_i)-\tilde p_i(x_i)|$

The statistics is over 30 random samples

```julia
replace!(ovls_avg, 1.0 => NaN);
```

```julia
using Plots

pl1 = scatter(ds, abs.(1 .-ovls_avg), yerr=ovls_std, xlabel="d", ylabel="|1 - overlap|",
    label="", yaxis=:log10)
vline!(pl1, [m], ls=:dash, label="no trunc", c=:gray, la=0.5)
pl2 = scatter(ds, errs_avg, yerr=errs_std, xlabel="d", ylabel="error on marginals", label="", yaxis=:log10)
vline!(pl2, [m], ls=:dash, label="no trunc", c=:gray, la=0.5)
pl = plot(pl1, pl2, size=(900,300), margin=5Plots.mm, legend=:topright)
```

## Incorporate into BP

```julia
using TensorTrains, Random, Tullio, TensorCast
using LinearAlgebra
using ProgressMeter
import TensorTrains.UniformTensorTrains: UniformTensorTrain

J = 0.1
β = 1.0
h = 0.2
function f(J, h, β, xᵢᵗ⁺¹,xₙᵢᵗ,xᵢᵗ)
    @assert xᵢᵗ⁺¹ ∈ 1:2
    @assert all(x ∈ 1:2 for x in xₙᵢᵗ)
    potts2spin(x) = 3-2x
    hⱼᵢ = β*J * sum(potts2spin, xₙᵢᵗ; init=0.0)
    E = - potts2spin(xᵢᵗ⁺¹) * (hⱼᵢ + β*h)
    return 1 / (1 + exp(2E))
end
```

```julia
function iterate_bp_vumps(f::Function, sz::Integer;
        maxiter=50, tol=1e-3,
        A0 = ones(sz, sz, 2, 2) .+ 1e-5 .* rand.())
    errs = fill(NaN, maxiter)
    ovls = fill(NaN, maxiter)
    εs = fill(NaN, maxiter)
    A = copy(A0)
    marg = fill(1/4, 4)
    @showprogress for it in 1:maxiter
        @tullio BB[m1,m2,n1,n2,xᵢᵗ,xⱼᵗ,xᵢᵗ⁺¹] := 
            f(J,h,β,xᵢᵗ⁺¹,[xⱼᵗ,xₖᵗ,xₗᵗ],xᵢᵗ)*A[m1,n1,xₖᵗ,xᵢᵗ]*A[m2,n2,xₗᵗ,xᵢᵗ] (xⱼᵗ in 1:2, xᵢᵗ⁺¹ in 1:2)
        @cast Q[(xᵢᵗ, xⱼᵗ, m1, m2), (n1, n2, xᵢᵗ⁺¹)] := BB[m1,m2,n1,n2,xᵢᵗ,xⱼᵗ,xᵢᵗ⁺¹]
        U, λ, V = svd(Matrix(Q))
        @cast C[m,k,xᵢᵗ,xⱼᵗ] := U[(xᵢᵗ, xⱼᵗ, m), k] k:length(λ), xᵢᵗ:2, xⱼᵗ:2
        @cast Vt[m,n,xᵢᵗ⁺¹] := V'[m, (n, xᵢᵗ⁺¹)]  xᵢᵗ⁺¹:2
        @tullio M[m,n,xᵢᵗ,xⱼᵗ] := λ[m] * Vt[m,l,xᵢᵗ] * C[l,n,xᵢᵗ,xⱼᵗ]
        
        Mresh = reshape(M, size(M,1), size(M,2), :)
        p = TensorTrains.InfiniteUniformTensorTrain(Mresh)
        Mtrunc, ovls[it] = truncate_vumps(complex.(permutedims(Mresh, (1,3,2))), sz)
        Mtrunc_resh = permutedims(Mtrunc, (1,3,2))
        q = TensorTrains.InfiniteUniformTensorTrain(Mtrunc_resh)
        marg_new = marginals(q) |> only
 
        εs[it] = maximum(abs, marg - only(marginals(p)))
        errs[it] = maximum(abs, only(marginals(p)) - only(marginals(q)))
        marg = marg_new
        A = reshape(Mtrunc_resh, size(Mtrunc_resh, 1), size(Mtrunc_resh, 2), 2, 2)
        εs[it] < tol && return A, maxiter, εs, errs, ovls
    end
    return A, maxiter, εs, errs, ovls
end
```

```julia
function belief(A, f::Function)
    @tullio BB[m1,m2,m3,n1,n2,n3,xᵢᵗ,xᵢᵗ⁺¹] := 
            f(J,h,β,xᵢᵗ⁺¹,[xⱼᵗ,xₖᵗ,xₗᵗ],xᵢᵗ)*A[m1,n1,xₖᵗ,xᵢᵗ]*A[m2,n2,xₗᵗ,xᵢᵗ]*A[m3,n3,xⱼᵗ,xᵢᵗ] (xᵢᵗ⁺¹ in 1:2)
    @cast M[(xᵢᵗ,m1,m2,m3), (n1,n2,n3,xᵢᵗ⁺¹)] := BB[m1,m2,m3,n1,n2,n3,xᵢᵗ,xᵢᵗ⁺¹]
    U, λ, V = svd(Matrix(M))
    @cast C[m,k,xᵢᵗ] := U[(xᵢᵗ,m),k] k:length(λ), xᵢᵗ:2
    @cast Vt[m,n,xᵢᵗ⁺¹] := V'[m, (n, xᵢᵗ⁺¹)]  xᵢᵗ⁺¹:2
    @tullio Anew[m,n,xᵢᵗ] := λ[m] * Vt[m,l,xᵢᵗ] * C[l,n,xᵢᵗ]
    Anew
end
```

```julia
d = 15
A, maxiter, εs, errs, ovls = iterate_bp_vumps(f, d; tol=1e-6, maxiter=40)
```

```julia
replace!(ovls, 1.0 => NaN);
p1 = plot(εs, xlabel="iter", ylabel="convergence error", label="")
p2 = plot(errs, xlabel="iter", ylabel="truncation error on marginals", label="")
p3 = plot(abs.(1 .- ovls), xlabel="iter", ylabel="|1-truncation overlap|", label="")
pl = plot(p1, p2, p3, yscale=:log10, layout=(1,3), size=(900,250), margin=5Plots.mm, labelfontsize=9)
```
